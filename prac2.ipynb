```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('temperatures.csv')
```

```python
df.head()
```

```python
x = df['YEAR']
y = df['JAN']
```

```python
plt.figure(figsize =(16,9))
plt.title('Temperature Plot of India')
plt.xlabel('Year')
plt.ylabel('Jan average Temperature')
plt.scatter(x,y)
```

```python
x
```

```python
x = x.values
```

```python
x = x.reshape(117,1)
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x,y)
```

```python
regressor.coef_
```

```python
regressor.intercept_
```

```python
regressor.predict([[2048]])
```

```python
predicted = regressor.predict(x)
```

```python
predicted
```

```python
import numpy as np
```

```python
#mean absolute error
np.mean(abs(y-predicted))
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y,predicted)
```

```python
#mean squared error
np.mean(abs(y - predicted)**2)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y, predicted)
```

```python
#r2 score
regressor.score(x,y)
```

```python
from sklearn.metrics import r2_score
r2_score(y,predicted)
```

```python
plt.title('Temperature plot of India')
plt.xlabel('Year')
plt.ylabel('Jan average Temp')
plt.scatter(x,y, label ='actual', color ='r')
plt.plot(x,predicted , label = 'predicted' , color = 'b')
plt.legend()
```
